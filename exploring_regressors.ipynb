```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
regressors = pd.read_csv('/BICNAS2/tuominen/trsnrm/regressors/Ratings_Grouped_GPT_Social_Aspects.csv')
```

```python
regressors.columns
```

```python

reg_mean_centered  = regressors.iloc[:,1::] - regressors.iloc[:,1::].mean()
stops= [(0,2),(3,4),(5,6)]

fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(16,18), sharex=True)
for i in range(len(stops)):
    sns.lineplot(data=reg_mean_centered.iloc[:,stops[i][0]:stops[i][1]+1], ax=ax[i])
    ax[i].set_title(f'Regressors {stops[i][0]} to {stops[i][1]}', fontsize=16)
    ax[i].set_ylabel('Mean Centered Rating', fontsize=14)
    ax[i].set_xlabel('Timepoints', fontsize=14)
plt.tight_layout()
plt.show()


    
```
